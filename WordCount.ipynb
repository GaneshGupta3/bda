### Theory
Word Count is a fundamental problem in text processing used to determine the frequency of words in a given input. It is widely used in applications like search engines, natural language processing (NLP), and big data analytics.

The concept is based on the **MapReduce paradigm**:
- **Map Phase:** Break input into smaller chunks and map each word as (word, 1)
- **Reduce Phase:** Aggregate counts of the same word

In this project, we simulate **parallel processing** by dividing the input into **3 blocks**, processing them independently, and then combining results. This mimics distributed systems like Hadoop.

# STEP 1: Take Input

We first take a sentence or paragraph as input from the user. This acts as the dataset for the word count process.

```python
text = input("Enter a sentence or paragraph:\n")
print("\nInput Text:\n", text)
```

# STEP 2: Preprocessing

We convert text to lowercase and remove punctuation. This ensures uniform counting so that words like "Word" and "word" are treated as the same.

```python
import string

text = text.lower()

# Remove punctuation
for p in string.punctuation:
    text = text.replace(p, "")

print("Cleaned Text:\n", text)
```

# STEP 3: Split into Words

The cleaned text is split into individual words which will be used for further processing.

```python
words = text.split()
print("Words List:\n", words)
```

# STEP 4: Divide into 3 Blocks (Parallel Simulation)

To simulate parallel processing, the list of words is divided into 3 parts. Each part represents a separate processor working independently.

```python
# Divide into 3 blocks
n = len(words)
block1 = words[:n//3]
block2 = words[n//3:2*n//3]
block3 = words[2*n//3:]

print("Block 1:", block1)
print("Block 2:", block2)
print("Block 3:", block3)
```

# STEP 5: Map Function

Each block performs the Map step, converting words into (word, 1) pairs. This represents the occurrence of each word.

```python
def map_function(block):
    mapped = []
    for word in block:
        mapped.append((word, 1))
    return mapped

map1 = map_function(block1)
map2 = map_function(block2)
map3 = map_function(block3)

print("Map Output Block 1:", map1)
print("Map Output Block 2:", map2)
print("Map Output Block 3:", map3)
```

# STEP 6: Shuffle and Combine

All mapped outputs from different blocks are combined together. This step groups all intermediate results before reduction.

```python
combined = map1 + map2 + map3
print("Combined Map Output:\n", combined)
```

# STEP 7: Reduce Function

In the Reduce phase, counts of each word are aggregated to produce the final frequency of each word.

```python
word_count = {}

for word, count in combined:
    if word in word_count:
        word_count[word] += count
    else:
        word_count[word] = count

print("\nFinal Word Count:")
for word, count in word_count.items():
    print(word, ":", count)
```

# Conclusion

This  demonstrates how word count can be efficiently performed using the MapReduce approach. By dividing the input into multiple blocks, we simulated parallel processing, which improves performance for large datasets. The use of mapping and reducing helps in organizing and aggregating data effectively. This approach is widely used in Big Data Analytics systems to handle large volumes of data. Hence, the implementation provides a basic understanding of distributed data processing and its importance in modern computing.
