```python
# Take input from user
text = input("Enter a sentence or paragraph:\n")

# Convert text to lowercase (optional, for uniform counting)
text = text.lower()

# Split text into words
words = text.split()

# Create dictionary to store word counts
word_count = {}

# MAP STEP:
# Here, each word is processed and mapped as (word, 1)
for word in words:
    # Remove punctuation (basic cleaning)
    word = word.strip(".,!?;:\"'()[]{}")

    # REDUCE STEP:
    # Aggregating counts for each word
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Display results
print("\nWord Count Result:")
for word, count in word_count.items():
    print(word, ":", count)
```
